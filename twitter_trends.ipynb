# Twitter Web Scraping Test

```python
# import dependencies 
from splinter import Browser
from bs4 import BeautifulSoup
import requests
import os
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://twitter.com/home/'
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup.prettify()
```

```python
soup.find_all('p')
```

```python
soup.find_all('p')[1].get_text()
```

# Web Scraping Using Tweepy and Twitter RESTful API 

```python
from tweetkey import consumer_api_key, consumer_api_secret_key, access_token, access_secret_token
```

```python
import tweepy
```

```python
auth = tweepy.OAuthHandler(consumer_api_key, consumer_api_secret_key)
auth.set_access_token(access_token, access_secret_token)

api = tweepy.API(auth)

public_tweets = api.home_timeline(200)
for tweet in public_tweets:
    print(tweet.text)
```
