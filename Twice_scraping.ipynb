```python
# Import tweetkey and other necessary dependencies
from tweetkey import consumer_api_key, consumer_api_secret_key, access_token, access_secret_token

import tweepy
import pandas as pd
import time 
```

```python
# Authorize the tweepy with OAuthHandler method from tweepy documentation 
# Set up my access token
auth = tweepy.OAuthHandler(consumer_api_key, consumer_api_secret_key)
auth.set_access_token(access_token, access_secret_token)

api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
# Search the query by the artist group name and show the results for popular and recent ones up to 150 items

twice_tweets = tweepy.Cursor(api.search, q='TWICE', result_type='mixed').items(150)
    
twice_tweets_list = [[twice_tweet.created_at, twice_tweet.id, twice_tweet.text, 
                      twice_tweet.favorite_count] 
                     for twice_tweet in twice_tweets
                    ]
    
twice_tweets_df = pd.DataFrame(twice_tweets_list)
```

```python
# Show the list of the tweets
twice_tweets_df
```

```python
cleaned_twice_tweets_df = twice_tweets_df.drop([1], axis = 1)
# cleaned_twice_tweets_df

new_twice_tweets_df = cleaned_twice_tweets_df.rename(columns={0: "Tweeted Date", 2: "Tweet Description", 3: "Popularity Count"})
# new_twice_tweets_df

sort_twice_tweets_df = new_twice_tweets_df.sort_values(by="Popularity Count", ascending=False)
sort_twice_tweets_df
```
